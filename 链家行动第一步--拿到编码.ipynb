```python
#从东城开始
#爬虫不需要反扒
dataall=pd.DataFrame({"NAME" : ["1","1"] ,"number":["1","1"]})
import requests 
import pandas as pd
import re
text=requests.get("https://bj.lianjia.com/ershoufang/dongcheng/pg2/")
textstart="https://bj.lianjia.com/ershoufang/dongcheng/pg"
textend="/"
textmid=""
for i in range(1,54,1):
    i = str(i)
    textmid=i
    d = textstart+textmid+textend
    text=requests.get(d)
    pattern = re.compile(r' data-el="region">(.*?) </a>')
    mytext = re.findall(pattern,text.text)
    mytext=pd.DataFrame(mytext)
    pattern1 = re.compile(r'class="btn-follow followBtn" data-hid="(.*?)"><span')
    mytext1 = re.findall(pattern1,text.text)
    mytext1=pd.DataFrame(mytext1)
    mytext["fangjia"]=mytext1.iloc[:,0]
    dataall=[dataall,mytext]
    dataall=pd.concat(dataall)
    print(d)
```
